```python
#Importing Libraries
import pandas as pd
import openpyxl
import xlrd
import xlwt
from xlwt.Workbook import *
#from pd import ExcelWriter
import xlsxwriter

#Reading the Excel file, Change the file name in below code if needed
dataframe = pd.read_excel('savedrecs.xls')

#Removing other columns and keeping data which is required
dataframe['Open Access Designations'] = dataframe[
  'Open Access Designations'].fillna('Paywalled_Journal')
col_list = [
  'Publication Type', 'Authors', 'Author Full Names', 'Article Title',
  'Affiliations', 'DOI', 'Open Access Designations'
]
dataframe = dataframe[col_list]
tempOA = dataframe['Open Access Designations']
tempOA = tempOA.fillna('Paywalled_Journal')

#Splitting the tempOA dataframe with ', '
for i in tempOA:
  x = i.split(', ')

#Reading the dataframe and counting number of rows in dataframe
print("Total Number of Journals: " + str(len(dataframe.index)))
print("\n")

#Creating List of OA designations - This list will be used to compare records
sub_list1 = [
  'gold', 'hybrid', 'Green Published', 'Green Accepted', 'Green Submitted',
  'Bronze', 'Paywalled_Journal'
]

#Initializing the variables for each OA designations
GreenAccepted = 0
Pay_Journal = 0
GreenPublished = 0
gold = 0
hybrid = 0
Green_submitted = 0
bronze = 0
result = {k: [] for k in sub_list1}

#The below for loop will count the number of records and store them in different sublist
for idx, row in dataframe.iterrows():
  if sub_list1[0] in row['Open Access Designations']:
    gold += 1
    result[sub_list1[0]].append(row.to_list())
    continue
  if sub_list1[1] in row['Open Access Designations']:
    hybrid += 1
    result[sub_list1[1]].append(row.to_list())
    continue
  if sub_list1[3] in row['Open Access Designations']:
    GreenAccepted += 1
    result[sub_list1[3]].append(row.to_list())
    continue
  if sub_list1[2] in row['Open Access Designations']:
    GreenPublished += 1
    result[sub_list1[2]].append(row.to_list())
    continue
  if sub_list1[4] in row['Open Access Designations']:
    Green_submitted += 1
    result[sub_list1[4]].append(row.to_list())
    continue
  if sub_list1[5] in row['Open Access Designations']:
    bronze += 1
    result[sub_list1[5]].append(row.to_list())
    continue
  if sub_list1[6] in row['Open Access Designations']:
    Pay_Journal += 1
    result[sub_list1[6]].append(row.to_list())

#Printing the number of Open Access Records
print("Gold Journals = " + str(gold))
print("Hybrid Journals =  " + str(hybrid))
print("Green Published Journals = " + str(GreenPublished))
print("Green Accepted Journals = " + str(GreenAccepted))
print("Green Submitted Journals = " + str(Green_submitted))
print("bronze Journals = " + str(bronze))
print("Paywalled Journals = " + str(Pay_Journal))

#Code for pie chart ( This code will not work on replit.com)
# print("\n")
# print("Pie Chart:")
# df_pie = pd.DataFrame({'Type': ['gold', 'hybrid','Green Published','Green Accepted',
#                         'Paywalled_Journal','Other'],
#                'count': [gold, hybrid, GreenPublished, GreenAccepted,Pay_Journal,Green_submitted+
#                         bronze]})
# df_pie.groupby(['Type']).plot(
#   kind = 'pie', y='count')

#Writing those sublist and creating different spreadsheets in single Excel File
writer = pd.ExcelWriter('New_File.xlsx', engine='xlsxwriter')
for i in sub_list1:
  temp_df = pd.DataFrame(result[i], columns=dataframe.columns)
  temp_df.to_excel(writer, f"{i}")
writer.save()
```
